```python
import lightkurve as lk
import pandas as pd
import os
import matplotlib.pyplot as plt
from read_tic import Create_Lightcurves
%matplotlib inline
import numpy as np
from scipy import signal
```

```python
if not os.path.exists('image'):
    os.makedirs('image')
path = os.path.dirname(os.path.abspath("lightkurve_tutorial_6.ipynb"))
```

```python
create_lightcurves = Create_Lightcurves()
tic_ids = create_lightcurves.read_tic_ids_from_pickle()
tic_ids_info=create_lightcurves.read_tic_ids_info_from_pickle()
n_tics = 0
batch_size=10
print(len(tic_ids))
```

```python
### Run from here--------------
```

```python
fig = plt.figure(figsize=(1024, 2048))    
fig, axs= plt.subplots(nrows = batch_size, ncols = 3, figsize=(20,15))
```

```python

k=n_tics
for i in range(batch_size):
    try:
        print(n_tics, i, batch_size, k)
        
        axs[i,0].set_title(tic_ids[n_tics]+": 1")
        axs[i,1].set_title(tic_ids[n_tics]+": 2")
        axs[i,2].set_title(tic_ids[n_tics]+": 3")
        
        search_tic_id = lk.search_lightcurve(tic_ids[n_tics], author='SPOC')
        tic_collection= search_tic_id.download_all()
        tic = tic_collection.stitch()
        tic_clean=tic.remove_outliers(sigma=20, sigma_upper=4)
        id_info = tic_ids_info[tic_ids_info['TIC ID'] == tic_ids[n_tics]].reset_index()
        period, epoch, duration_hours = id_info['Period (days)'][0], id_info['Epoch (BTJD)'][0], id_info['Duration (hours)'][0]
        lc_folded = tic_clean.fold(period= period, epoch_time=epoch)
        #fractional_duration = (duration_hours / 24.0) / period
        fractional_duration = (duration_hours / 24.0)
        phase_mask = (lc_folded.phase > -3.0*fractional_duration) & (lc_folded.phase < 3.0*fractional_duration)
        lc_zoom = lc_folded[phase_mask]
        #########################
        binned_lc_1=lc_zoom.bin(time_bin_size=(fractional_duration*0.16)/101.0).normalize()-1
        binned_lc_1=(binned_lc_1 / np.abs(np.nanmin(binned_lc_1.flux)) ) * 2.0 + 1
        binned_lc_1=binned_lc_1.to_pandas().reset_index()
        ######################### 
        binned_lc_2=lc_zoom.bin(bins=301).normalize()-1
        binned_lc_2=(binned_lc_2 / np.abs(np.nanmin(binned_lc_2.flux)) ) * 2.0 + 1
        binned_lc_2=binned_lc_2.to_pandas().reset_index()
        #########################
        #binned_lc_3=lc_zoom.bin(time_bin_size=(duration_hours*0.16)/(24.0*301), n_bins=301).normalize()-1
        phase_mask = (lc_folded.phase > -1.0*fractional_duration)& (lc_folded.phase < 1.0*fractional_duration)
        lc_zoom = lc_folded[phase_mask]
        binned_lc_3= lc_zoom.bin(time_bin_size=fractional_duration/150).normalize() - 1
        binned_lc_3 = (binned_lc_3 / np.abs(np.nanmin(binned_lc_3.flux)) ) * 2.0 + 1
        binned_lc_3=binned_lc_3.to_pandas().reset_index()

        x = np.array(binned_lc_1.time)
        y = np.array(binned_lc_1.flux)
        axs[i,0].scatter(x,y, s=5, c='#004444')
        
        x = np.array(binned_lc_2.time)
        y = np.array(binned_lc_2.flux)
        axs[i,1].scatter(x,y, s=5, c='black')
        
        x = np.array(binned_lc_3.time)
        y = np.array(binned_lc_3.flux)
        axs[i,2].scatter(x,y, s=5, c='#888888')
        
    
        n_tics+=1
        
    except Exception as e:
        print(f"Exception Raised: {e}\n\n"f"Validate TIC ID:{tic_ids[n_tics]}")
        n_tics+=1
        continue
    
    
    
```

```python
fig.tight_layout(pad=1.0)   
fig.savefig(path+f"\\image\\image_{k}_{k+batch_size-1}.png", bbox_inches ="tight", orientation ='landscape')
```
