```python
import lightkurve as lk
import pandas as pd
import matplotlib.pyplot as plt
from read_tic import Create_Lightcurves
%matplotlib inline
import numpy as np
from scipy import signal
```

```python
create_lightcurves = Create_Lightcurves()
tic_ids = create_lightcurves.read_tic_ids_from_pickle()
print(len(tic_ids))

tic_ids_info=create_lightcurves.read_tic_ids_info_from_pickle()
```

```python
search_tic_id = lk.search_lightcurve(tic_ids[1], author=('SPOC', 'TESS-SPOC'))
```

```python
tic_collection= search_tic_id.download_all()
```

```python
tic_collection.plot()
```

```python
tic = tic_collection.stitch()
```

```python
id_info = tic_ids_info[tic_ids_info['TIC ID'] == tic_ids[1]].reset_index()
```

```python
lc = tic.fold(period= id_info['Period (days)'][0], epoch_time=id_info['Epoch (BTJD)'][0]).normalize().remove_outliers()
ax = lc.scatter()
ax.set_xlim(-0.5, 0.5);
```

```python
lc = lc.to_pandas().reset_index()
lc_folded = lc[['time', 'flux']]
lc_folded.shape
```

```python
x_detrended = signal.detrend(lc_folded[:,1])
plt.figure(figsize=(15,8))
plt.scatter(lc['time'], lc['flux'], label="x")
#plt.scatter(lc['time'], x_detrended, label="x_detrended")
plt.legend(loc='best')
plt.xlim(-0.5, 0.5)
plt.show()
```

```python
# x_detrended = signal.detrend(np.array(lc['flux']))
print(x_detrended.shape)
plt.figure(figsize=(15,8))
#plt.scatter(lc['time'], lc['flux'], label="x")
plt.scatter(lc_folded[:,0], x_detrended, label="x_detrended")
plt.legend(loc='best')
plt.xlim(-0.5, 0.5)
plt.show()
```
